## Exploration

```python
# Libraries
import pandas as pd
from pprint import pprint
```

```python
def load_yelp_df(table_name: str, nrows=None):
    chunks = pd.read_json(f'data/yelp_academic_dataset_{table_name}.json', lines=True, chunksize=100000, nrows=nrows)
    return pd.concat(chunks)
```

```python
# Business table
business_df = load_yelp_df('business', nrows=5)
pprint(list(business_df.columns))
business_df.head()
```

```python
# Checkin table
checkin_df = load_yelp_df('checkin', nrows=5)
pprint(list(checkin_df.columns))
checkin_df.head()
```

```python
# Review table
review_df = load_yelp_df('review', nrows=5)
pprint(list(review_df.columns))
review_df.head()
```

```python
# Tip table
tip_df = load_yelp_df('tip', nrows=5)
pprint(list(tip_df.columns))
tip_df.head()
```

```python
# User table
user_df = load_yelp_df('user', nrows=5)
pprint(list(user_df.columns))
user_df.head()
```
